```python
from __future__ import print_function
import os
import matplotlib.pyplot as plt
import pylab as py
import re
import pickle
import numpy as np
%matplotlib inline
```

```python
path2results='sidis_hermes_torino_alexei_dy'
# out
fig_kt_dy = "kt_dy_alexei.pdf"
fig_pt_pion_dy = "pt_pion_dy_alexei.pdf"
fig_pt_kaon_dy = "pt_kaon_dy_alexei.pdf"
fig_gk0_dy = "gk0_dy_alexei.pdf"
fig_Q0_dy = "Q0_dy_alexei.pdf"
```

```python
dystat = {'125':1.25,'15':1.5,'175':1.75,'2':2.,'225':2.25,'25':2.5,'275':2.75,'3':3.,'325':3.25,'35':3.5}
```

```python
dy = np.zeros(len(dystat))
val = np.zeros(len(dystat))
valerr = np.zeros(len(dystat))
sea = np.zeros(len(dystat))
seaerr = np.zeros(len(dystat))
fav = np.zeros(len(dystat))
faverr = np.zeros(len(dystat))
unfav = np.zeros(len(dystat))
unfaverr = np.zeros(len(dystat))
kfav = np.zeros(len(dystat))
kfaverr = np.zeros(len(dystat))
kunfav = np.zeros(len(dystat))
kunfaverr = np.zeros(len(dystat))
gk0 = np.zeros(len(dystat))
gk0err = np.zeros(len(dystat))
Q0 = np.zeros(len(dystat))
Q0err = np.zeros(len(dystat))




i=0
for prefix in dystat.keys():
    with open(path2results+'/descriptive_alexei_dy'+prefix+'.pkl', 'rb') as handle:
        stat = pickle.load(handle)
        dy[i] = stat['val'][0]
        val[i] = stat['val'][1]
        valerr[i] = stat['val'][2]
        sea[i] = stat['sea'][1]
        seaerr[i] = stat['sea'][2]
        fav[i] = stat['fav'][1]
        faverr[i] = stat['fav'][2]
        unfav[i] = stat['unfav'][1]
        unfaverr[i] = stat['unfav'][2]
        kfav[i] = stat['kfav'][1]
        kfaverr[i] = stat['kfav'][2]
        kunfav[i] = stat['kunfav'][1]
        kunfaverr[i] = stat['kunfav'][2]
        gk0[i] = stat['gk0'][1]
        gk0err[i] = stat['gk0'][2]
        Q0[i] = stat['Q0'][1]
        Q0err[i] = stat['Q0'][2]
        i=i+1
```

```python
dy
```

```python
fig, ax = plt.subplots()
ax.set_ylabel(r"$\langle k_\perp^2 \rangle_{} $ (GeV$^2$)", size=14)
ax.set_xlabel("dy",labelpad=1)

ax.errorbar(dy, val, valerr, capsize=5, fmt="o",color="blue")
ax.errorbar(dy, sea, seaerr, capsize=5, fmt="o",color="red")

ax.legend(["valence", "sea"], loc="lower center", ncol=2, fontsize="large")
py.savefig("%s/%s" % (path2results, fig_kt_dy))
```

```python
fig, ax = plt.subplots()
ax.set_ylabel(r"$\langle p_\perp^2 \rangle_{\pi} $ (GeV$^2$)", size=14)
ax.set_xlabel("dy",labelpad=1)

ax.errorbar(dy, fav, yerr=faverr, capsize=5, fmt="o",color="blue")
ax.errorbar(dy, unfav, yerr=unfaverr, capsize=5, fmt="o",color="red")

ax.legend(["fav", "unfav"], loc="lower center", ncol=2, fontsize="large")
py.savefig("%s/%s" % (path2results, fig_pt_pion_dy))
```

```python
fig, ax = plt.subplots()
ax.set_ylabel(r"$\langle p_\perp^2 \rangle_{K} $ (GeV$^2$)", size=14)
ax.set_xlabel("dy",labelpad=1)

ax.errorbar(dy, kfav, yerr=kfaverr, capsize=5, fmt="o",color="blue")
ax.errorbar(dy, kunfav, yerr=kunfaverr, capsize=5, fmt="o",color="red")

ax.legend(["fav", "unfav"], loc="lower center", ncol=2, fontsize="large")
py.savefig("%s/%s" % (path2results, fig_pt_kaon_dy))
```

```python
fig, ax = plt.subplots()
ax.set_ylabel(r"$g_{K0} $", size=14)
ax.set_xlabel("dy",labelpad=1)

ax.errorbar(dy, gk0, yerr=gk0err, capsize=5, fmt="o",color="red")

py.savefig("%s/%s" % (path2results, fig_gk0_dy))
```

```python
fig, ax = plt.subplots()
ax.set_ylabel(r"$Q_{0} $ (GeV)", size=14)
ax.set_xlabel("dy",labelpad=1)

ax.errorbar(dy, Q0, yerr=Q0err, capsize=5, fmt="o",color="red")

py.savefig("%s/%s" % (path2results, fig_Q0_dy))
```
